```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
degrees = np.arange(0,361,30)
radian_values = [f"{deg/180:.2f}" for deg in degrees] 
print(radian_values)
```

```python
x = np.arange(1,21,1)
y = np.arange(10,210,10)
plt.figure(figsize=(7,7))
plt.plot(x,y, 'g-*')
plt.show()
```

```python
x_sin = np.arange(0,2*np.pi,0.1)
y_sin = np.sin(x_sin)

print(y_sin)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_sin,y_sin,'r-*')
plt.title('Sign Curve')
plt.show()
```

```python
x_cos = np.arange(0,2*np.pi,0.1)
y_cos = np.cos(x_cos)
plt.figure(figsize=(6,6))
plt.plot(x_cos,y_cos,'g-*')
plt.title('Cos Curve')
plt.show()
```

```python
x_tan = np.arange(0,2*np.pi,0.1)
y_tan = np.tan(x_tan)
plt.figure(figsize=(6,6))
plt.plot(x_tan,y_tan)
plt.title('Tan Curve')
plt.show()
```

```python
x_cot = np.arange(0,2*np.pi,0.1)
y_cot = 1/np.tan(x_cot)
plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot)
plt.title('Cot Curve')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.subplot(2,2,1)
plt.plot(x_sin,y_sin,'r-')
plt.title('Sign Curve')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,'b-')
plt.title('Cos Curve')

plt.subplot(2,2,3)
plt.plot(x_tan,y_tan,'g-')
plt.title('Tan Curve')

plt.subplot(2,2,4)
plt.plot(x_cot,y_cot,'-')
plt.title('Cot Curve')

plt.show()
```
